```python
import pandas as pd
import numpy as np
import os
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('scopus.csv')
df.head()
```

```python
# Replace keywords from abstracts
df['Abstract'] = df['Abstract'].apply(lambda x: 
                                          x.replace('BACKGROUND:','').replace('BACKGROUNDS:','').replace('Backgroud:','')
                                          .replace('INTRODUCTION:','').replace('Introduction:','')
                                          .replace('OBJECTIVES:','').replace('OBJECTIVE:','').replace('METHODS:','').replace('METHOD:','')
                                          .replace('RESULTS:','').replace('RESULT:','')
                                          .replace('CONCLUSION:','').replace('CONCLUSIONS:',''))
# Convert the abstract and the title to lowercase
df['Abstract'] = df['Abstract'].apply(lambda x: x.lower())
#Convert the title to lowercase
df['Title'] = df['Title'].apply(lambda x: str(x).lower())
#Search terms related to health and kept row that include them
df = df[((df['Abstract'].str.contains(' health| disease'))|
         (df['Title'].str.contains(' health| disease')))]
```

```python
# Detect and remove articles with abstracts in foreign language
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0
def langdet (x):
    try:
        return detect(x)
    except:
        return "NA"
df['lang'] = df['Abstract'].apply(lambda x: langdet(x))
df = df[df['lang'].str.contains('en')]
```

```python
# Convert the clean abstract to tokens and find the unigrams and bigrams
import re
import nltk
import string
from nltk import bigrams
stopword = nltk.corpus.stopwords.words('english')
def tokenization(text):
    text = text.split()
    text = ','.join(set(text))
    return text
def clean_text(text):
    text_lc = " ".join([word.lower() for word in text.split() if word not in string.punctuation]) 
    text_rc = re.sub('[0-9]+', '', text_lc)
    tokens = re.split('\W+', text_rc)   
    text = [word for word in tokens if word not in stopword]  
    text = ' '.join(text)
    return text
df['clean_text'] = df['Abstract'].apply(lambda x: clean_text(x))
df['unigram'] = df['clean_text'].apply(lambda x: tokenization(x))
df['bigram']  = df['unigram'].apply(lambda x: ','.join([st[0].strip()+" "+st[1].strip() for st in list(bigrams(x.split(',')))]))
```

```python
# Use TfidfVectorizer for feature engineering
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
import re
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")
def tokenize_and_stem(text):
    tokens = [word for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems
# Create Tfid matrix 
tfidf_vectorizer = TfidfVectorizer(max_df=0.90, min_df=10, stop_words='english',
                                 use_idf=True, tokenizer=tokenize_and_stem, ngram_range=(1,2))

tfidf_matrix = tfidf_vectorizer.fit_transform(df['clean_text'].tolist()) 
print(tfidf_matrix.shape)
```

```python
# Search best number of clusters using Elbow method
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
from sklearn.decomposition import TruncatedSVD

pca = TruncatedSVD(n_components=100, random_state=33)
X = pca.fit_transform(tfidf_matrix)

# Iniciamos el modelo de clusters y visualizamos
model = KMeans(init = 'k-means++', random_state=33)
visualizer = KElbowVisualizer(model, k=(2,10))
visualizer.fit(X)      
visualizer.show()
```

```python
# We use K_value as input for K means clustering
K_value = 4
X = tfidf_matrix
from sklearn.cluster import KMeans
km = KMeans(n_clusters = K_value, init = 'k-means++', random_state = 0)
fitted = km.fit(X)
predict = km.predict(X)
```

```python
# Assign number of clusters to each article
df['cluster'] = pd.Series(predict, index = df.index)
```
